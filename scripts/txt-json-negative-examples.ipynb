```python
import json

input_path = "/kaggle/input/negative-examples/negative_examples.txt"
output_path = "/kaggle/working/negative_examples.json"

with open(input_path, 'r', encoding='utf-8') as infile, \
     open(output_path, 'w', encoding='utf-8') as outfile:
    for line in infile:
        clean_line = line.strip()
        if clean_line:  # Ignora linhas vazias
            entry = {
                "instruction": line, 
                "input": "", 
                "output": "I can't answer that."
            }
            outfile.write(json.dumps(entry, ensure_ascii=False) + ', \n')
```
